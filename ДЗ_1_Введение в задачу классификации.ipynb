### Домашнее задание 

###### 1.Приведите по 2 примера, когда лучше максимизировать Precision, а когда Recall.

Precision - лучше максимизировать тогда, когда важна строгость в определении результата. К примеру при кредитной политики банка -
для уменьшения риска по не возврату кредитов производится более строгая оценка платёжеспособности заёмщиков, даже если это ведет
к риску потери части потенциально платежеспособных клиентов.

Также Precision стоит максимизировать тогда, когда в бане используется модель для определения мошеннических транзакции.
В этом случае лучше заблокировать операцию до выяснения, чем допустить возможную опасную ситуацию.

Recall - лучшее максимизировать тогда, когда лучшее переоценить чем недооценить. Примером может являться диагностика пациентов.
Когда нельзя не выявить заболевание.

Также Recall стоит максимизировать тогда, когда политика банка такова - выдача большего количества кредитов, даже если риск
не возврата увеличивается.

###### 2. Почему мы используем F-меру, почему, например, нельзя просто взять среднее от Precision и Recall?

F-score - метрика объединяющая в себе точность и полноту, выдавая как результат их гармоническое среднее.
С помощью коэффициента β, мы можем в зависимости от задачи повышать или понижать приоритет Precision или Recall,
максимизируя то, что для нас наиболее приоритетно.
